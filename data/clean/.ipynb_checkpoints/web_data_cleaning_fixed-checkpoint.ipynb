```python
import pandas as pd 
```

```python
df1 = pd.read_csv("../raw/df_final_web_data_pt_1.txt", sep=",")
df2 = pd.read_csv("../raw/df_final_web_data_pt_2.txt", sep=",")
```

```python
df_web_data = pd.concat([df1, df2], axis=0, join="outer")
df_web_data.head()
```

```python
df_web_data.shape
```

```python
df_web_data["client_id"] = pd.to_numeric(df_web_data["client_id"], errors='coerce', downcast='integer')
df_web_data["client_id"] = df_web_data["client_id"].astype("Int64")
df_web_data["date_time"] = pd.to_datetime(df_web_data["date_time"])
```

```python
df_web_data.dtypes
```

```python
#Checking for null values 
df_web_data.isnull().values.any()
```

```python
#Total number of clients in the dataset (clients that did the process)

total_clients = df_web_data["client_id"].nunique()
print(f"Total number of clients in the dataset: {total_clients}")
```

```python
# Removing duplicates only for the 'start' and 'confirm' step
steps_to_dedup = ["start", "confirm"]

df_special = df_web_data[df_web_data["process_step"].isin(steps_to_dedup)]

df_special_last = df_special.sort_values("date_time") \
    .drop_duplicates(subset=["visit_id", "process_step"], keep="last")

df_other = df_web_data[~df_web_data["process_step"].isin(steps_to_dedup)]

df_final = pd.concat([df_special_last, df_other]) \
             .sort_values(["visit_id", "date_time"])
```
